```python
import nltk
import numpy as np
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
def convert_lower_case(data):
    return np.char.lower(data)

def remove_punctuation(data):           #remove Punctuations from text
    symbols = "!\"#$%&()*+-./:;<=>?@[\]^_`{|}~\n"
    for i in range(len(symbols)):
        data = np.char.replace(data, symbols[i], ' ')
        data = np.char.replace(data, "  ", " ")
    data = np.char.replace(data, ',', '')
    return data

def remove_apostrophe(data):            #Remove Apostrophe from text
    return np.char.replace(data, "'", "")

def remove_stop_words(data):          #Remove Stopwords from tokens
    stop_words = stopwords.words('english')
    words = word_tokenize(str(data))
    new_text = ""
    for w in words:
        if w not in stop_words:
            new_text = new_text + " " + w
    return np.char.strip(new_text)





def preprocessed(query):            # execute preprocessing
    query = convert_lower_case(query)
   
    query = remove_punctuation(query)
    query = remove_apostrophe(query)
    query = remove_stop_words(query)
    return query
```
